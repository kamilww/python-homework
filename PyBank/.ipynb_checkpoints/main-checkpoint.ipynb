```python
from pathlib import Path
import csv

total_months = 0
total = 0
average_change = 0
greatest_increase = 0
greatest_decrease = 0

# Set the file path
filepath = Path("budget_data/02-Homework_02-Python_Instructions_PyBank_Resources_budget_data.csv")

# Open the csv file as an object
with open(filepath, 'r') as csvfile:

    # Use DictReader to store csv data under csvdict
    csvdict = csv.DictReader(csvfile)
    
    # Iterating through the dict elements in csv count the number of months based on number of dataset entries
    for item in csvdict:
        total_months += 1
    
    # Iterating through each dict items in csvdict
    for item in csvdict:
        
        csv_line_item = item["Profit/Losses"]
        
        # count total profit/losses
        total += csv_line_item
         
        # Finding max/min values. If csv_line_item is zero, loop continues to next item
        if csv_line_item == 0:
            continue
        
        # find the largest negative number by absolute value and assign to greatest_decrease
        elif csv_line_item < 0 and csv_line_item < greatest_decrease:
            greatest_decrease = csv_line_item
        
        # find the largest positive value and assign to greatest_increase
        elif csv_line_item > greatest_increase:
            greatest_increase = csv_line_item
    
    # count average change
    average_change = round(total/total_months, 2)
            
            
            
csvfile.close()         
```

```python
from pathlib import Path
import csv

# Set the path for the output.csv
output_path = Path('output/output.txt')

# Open the output path as a file
with open(output_path, 'w') as txt_output:
    csv.writer(f"""
        Total Months: {total_months}
        Total: ${total:,.2f}
        Average Change: ${average_change}
        
    """)
    
txt_output.close()
    
```
