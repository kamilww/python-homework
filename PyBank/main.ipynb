```python
from pathlib import Path
import csv

# Set variables
total_months = 0
total = 0
average_change = 0
greatest_increase = 0
greatest_decrease = 0
greatest_increase_date = ""
greatest_decrease_date = ""

# Set the file path
filepath = Path('budget_data/02-Homework_02-Python_Instructions_PyBank_Resources_budget_data.csv')

# Open the csv file as an object
with open(filepath, 'r') as csvfile:

    # Use DictReader to store csv data under csvdict
    csvdict = csv.DictReader(csvfile)
    
    # Iterating through the dict elements
    for item in csvdict:
        
        # Retrieve P/L data from dict element item, casting as integer and storing element in loop variable csv_line_item
        csv_line_item = int(item['Profit/Losses'])
        
        # Count total entries and store in variable total_months
        total_months += 1 
        
        # Count total profit/losses, storing in total
        total += csv_line_item
         
        # Find the largest negative number by absolute value and assign to greatest_decrease
        # Save corresponding date entry of csv_line_item to greatest_decrease_date
        if csv_line_item < 0 and csv_line_item < greatest_decrease:
            greatest_decrease = csv_line_item
            greatest_decrease_date = item['Date']
        
        # Find the largest positive value and assign to greatest_increase
        # Save corresponding date entry of csv_line_item to great_increase_date
        elif csv_line_item > greatest_increase:
            greatest_increase = csv_line_item
            greatest_increase_date = item['Date']
        
    # count average change in data's P/L
    average_change = round(total/total_months, 2)
            
csvfile.close()         

# Set the path for the output.csv
output_path = Path('output/output.txt')

# Open the output path as a file
with open(output_path, 'w') as txt_output:
    
    # Pass in txt_output file into csv.writer():
    csvwriter = csv.writer(txt_output, delimiter='\n')
    
    # Store variables in string
    csvresults = ["Financial Analysis",
        "----------------------------",
        f"Total Months: {total_months}",
        f"Total: ${total}",
        f"Average Change: ${average_change}",
        f"Greatest Increase in Profits: {greatest_increase_date} ({greatest_increase})",
        f"Greatest Decrease in Profits: {greatest_decrease_date} ({greatest_decrease})"
        ]
    
    # Write csvresults to csv output file and print csvresults to console
    csvwriter.writerow(csvresults)
    
    for row in csvresults:
        print(row)
        
        
txt_output.close()
    
```
